```python
## POS tagging: part of speech tagging
nltk.download("averaged_perceptron_tagger")

## get tokens first
word_list = reuters.words("training/9793")
nltk.pos_tag(word_list)
```

```python
## Chunking
# Parsing using regular expression with chunking (without chinking)

# We specify that noun phrase can have a determinant, adverb, gerund verb, or an adjective,
# but it must have a noun or a pronoun, e.g. "A fastly running beautiful deer..."
# The verb phrase should start with a verb, and then it can have anything.

pattern = """NP: {<DT>?<RB.?>?<VBG>?<JJ.?>*(<NN.?>|<PRP.?>)+}
             VP: {<VB.?>+<.*>*}
"""

mySentence = "A fastly running beautiful deer skidded off the road"
## parsing the pattern with Regular expression parser to understand the pattern
myParser = nltk.RegexpParser(pattern)
## first get words of senetence .. then .. do pos_tagging... then only it will work .. else it will not work
myParsedSentence = myParser.parse(nltk.pos_tag(nltk.word_tokenize(mySentence)))
print(myParsedSentence)
```

```python
# Let us try another sentence with chunking (without chinking)

mySentence = 'I left to do my homework'

myParser = nltk.RegexpParser(pattern)
myParsedSentence = myParser.parse(nltk.pos_tag(nltk.word_tokenize(mySentence)))
print(myParsedSentence)
```

```python
# Chunking and Chinking
## In chinking .. }(<VBG>|(<TO><.*>*)){ brackets are in revrese order..

# Redefine pattern with chinking, to exclude "to do something"
pattern = """NP: {<DT>?<RB.?>?<VBG>?<JJ.?>*(<NN.?>|<PRP.?>)+}
             VP: {<VB.?>+<.*>*}
                 }(<VBG>|(<TO><.*>*)){
"""

mySentence = 'I left to do my homework'

myParser = nltk.RegexpParser(pattern)
myParsedSentence = myParser.parse(nltk.pos_tag(nltk.word_tokenize(mySentence)))
print(myParsedSentence)
```

```python
## context free grammner(cfg)
## declare own rule

from nltk import CFG

myGrammar = nltk.CFG.fromstring("""
    S -> NP VP
    VP -> VB NP
    VP -> VB
    VP -> VB PRP
    NP -> DET NN
    
    VB -> "chased"|"ate"
    DET -> "another"|"the"
    NN -> "cat"|"rat"|"snake"
    PRP -> "it"
""")

# Generating sentences from the defined grammar

from nltk.parse.generate import generate
for sent in generate(myGrammar):
    print(' '.join(sent))
```

### Stemming & Lemmatization
#### We will import one new corpus " brown "
#### we are not using reuters corpus as it contain alphanumeric words also , and here we are trying to understand 
#### the stemming and lemmatization .. It is not like that we cannt use reuters corpus but for better sake of understanding only we are #### using brown corpus

```python
nltk.download("brown")
from nltk.corpus import brown
```

```python
## get documents ids
brown.fileidsids()
```

```python
fileID = "ck23"

## remove stopwords
from nltk.corpus import stopwords
filter_words = [w for w in brown.words(fileID) if w.lower() not in stopwords.words("english")]

## remove punctuation
import string
filter_words = [w for w in filter_words if w not in string.punctuation]
```

```python
## Compare two stemmers 
## PorterStemmer : Old, less rule, fast.
## LancasterStemmer : New, More Rules, slow.

from nltk.stem import PorterStemmer
from nltk.stem import LancasterStemmer

porter = PorterStemmer()
lancaster = LancasterStemmer()

for word in filter_words:
    print("Actual World :{}, PorterStemmer Word :{} , LancasterStemmer Word:{}". format(word, porter.stem(word), lancaster.stem(word)))
    
```

```python
## Compare Lemmatization to Stemming
from nltk.stem import WordNetLemmatizer

wordlemmatizer = WordNetLemmatizer()

for word in filter_words:
    print("Actual World :{}, PorterStemmer Word :{} , LancasterStemmer Word:{}, Lemmatized Word:{}". format(word, porter.stem(word), lancaster.stem(word), wordlemmatizer.lemmatize(word)))
```
